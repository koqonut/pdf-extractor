```python
from pdf2image import convert_from_path
import os
```

```python
data_dir = '../data/'
flyer = 'metro'
flyer_filename = flyer + ".pdf"
```

```python
# Define the relative path to the PDF file
pdf_path = os.path.join(data_dir,'raw',flyer_filename)
```

```python
pages = convert_from_path(pdf_path, 300)
print("Number of pages: ", len(pages))
```

```python
# takes 38 seconds
for i, page in enumerate(pages):
    filename = f'pdf2image_page_{i + 1}.png'
    pdf_output = os.path.join(data_dir,'images',flyer,filename)
    page.save(pdf_output, 'PNG')
```
